# Data plots

This example illustrates data plotting for summary statistics of observed vs. simulated data. This can be used for easy assessment of how good we can fit the data.

Let’s start to import the necessary classes. We also set up matplotlib and we’re going to use numpy and pandas as well.

```python
from pyabc.visualization import data_plot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Define summary statistics for observed data

First, we will define some arbitrary summary statistics to be used in this example. We will define four different summary statistics with different data types. Data types of the value of the summary statistic can be 1d numpy array, 2d numpy array, dictionary.

```python
observed = {'data 1': np.array([1,1,2,2,3,3,4,4]),
            'data 2': pd.DataFrame([{'measurement': [1,2,3,4,1,2,3,4]}]),
            'data 3': np.array([[4,4,3,3,2,2,1,1],[4,4,3,3,2,2,1,1]]),
            'data 4': np.array([1,1,2,2,3,3,4,4])}
```

## Define summary statistics for simulated data

We will do the same for the simulated data. We will define four different summary statistics. 

```python
simulated = {'data 1': np.array([1,2,4,6,8,10,12,14]),
             'data 2': pd.DataFrame([{'measurement': [1,2,4,6,1,2,4,6]}]),
             'data 3': np.array([[14,12,10,8,6,4,2,1],[14,12,10,8,6,4,2,1]]),
             'data 4': np.array([1,2,4,6,8,10,12,14])}
```

Now that we have defined two dictionaries for both the observed and simulated summary statistics. we can call the plotting function from pyABC.visualization

```python
data_plot.plot_data(observed, simulated)
plt.show()
```
